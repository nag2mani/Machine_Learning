```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the input image and show its dimensions, keeping in mind that
# images are represented as a multi-dimensional NumPy array with
# shape no. rows (height) x no. columns (width) x no. channels (depth)
image = cv2.imread("img.jpeg", cv2.IMREAD_COLOR)
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

```python
# Convert BGR image to RGB (matplotlib uses RGB)
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
img1 = cv2.imread("img.jpeg" ,0)
img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
histogram = cv2.calcHist([img1], [0], None, [256], [0, 256])
cdf = histogram.cumsum()
plt.plot(histogram/max(histogram))
plt.plot(cdf/max(cdf))
plt.show()
```

```python
equ = cv2.equalizeHist(img1)
img_rgb = cv2.cvtColor(equ, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```python
histogram = cv2.calcHist([equ], [0], None, [256], [0, 256])
cdf = histogram.cumsum()
plt.plot(histogram/max(histogram))
plt.plot(cdf/max(cdf))
plt.show()
```
