```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# It is very disblance data.
train_df["similarity manual"].value_counts()
```

```python
X=train_df[["sbert", "pubmedbert", "biobert", "puborgbert", "bioorgbert", "bert", "llama", "m2bert", "bge"]]

y = train_df[["similarity manual"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {lr.score(X_train, y_train)}")
print(f"Testing Data Score: {lr.score(X_test, y_test)}")
```

```python
y_predict = lr.predict(X_train)
accuracy = metrics.accuracy_score(y_train, y_predict)
print("Training Accuracy : ",accuracy)
```

```python
y_test_pred = lr.predict(test_df)
```
