```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
Predicted_values_by_my_model = [5.680787126761226, 2.384060066057183] #w0, w1
Predicted_values_by_Sklearn_model = [5.68078713, 2.38406007]

Error_by_my_model = [1.2805559784291467, 2.0785254017773265, 1.4417091945941547, 0.9579571905586358] #mae, mse,rmse,r_square
Error_by_Sklearn_model = [1.280555978429147, 2.078525401777328, 1.4417091945941551, 0.9579571905586357]

# Data
Predicted_values_by_my_model = [5.680787126761226, 2.384060066057183]
Predicted_values_by_Sklearn_model = [5.68078713, 2.38406007]

Error_by_my_model = [1.2805559784291467, 2.0785254017773265, 1.4417091945941547, 0.9579571905586358]
Error_by_Sklearn_model = [1.280555978429147, 2.078525401777328, 1.4417091945941551, 0.9579571905586357]

# Bar graph for predicted values
labels = ['w0', 'w1']
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(np.arange(len(labels)) - width/2, Predicted_values_by_my_model, width, label='My Model')
rects2 = ax.bar(np.arange(len(labels)) + width/2, Predicted_values_by_Sklearn_model, width, label='Sklearn Model')

ax.set_ylabel('Values')
ax.set_title('Predicted Values by My Model and Sklearn Model')
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.legend()

# Display the bar graph for predicted values
plt.show()

# Bar graph for errors with horizontal line for R-Square
error_labels = ['MAE', 'MSE', 'RMSE', 'R-Square']
fig, ax = plt.subplots()
rects1 = ax.bar(np.arange(len(error_labels)) - width/2, Error_by_my_model, width, label='My Model')
rects2 = ax.bar(np.arange(len(error_labels)) + width/2, Error_by_Sklearn_model, width, label='Sklearn Model')

ax.set_ylabel('Error Values')
ax.set_title('Errors by My Model and Sklearn Model')
ax.set_xticks(np.arange(len(error_labels)))
ax.set_xticklabels(error_labels)
ax.legend()

# Add a horizontal line for R-Square
rsquare_value = 0.9579571905586358  # Change this value to the actual R-Square value
plt.axhline(y=rsquare_value, color='r', linestyle='--', label='R-Square')

# Display the bar graph for errors with the horizontal line for R-Square
plt.show()
```

# Gradient Descent v/s Stochastic Gradient Descent on Data4.csv 

```python
Predicted_values_by_my_model = [13.239477941118366, 6.132437615313393, 2.392265549258809, 7.746810380660236] #w0, w1, w2, w3
Predicted_values_by_GD = [13.18578203958556, 6.060680905255514, 2.445134785871812, 7.766764500756425]
Predicted_values_by_SGD = [13.154571488555078, 6.091250059460483, 2.3134524409841464, 7.827527898786717]


# Bar graph for predicted values
labels = ['w0', 'w1', 'w2', 'w3']
width = 0.30

fig, ax = plt.subplots()
rects1 = ax.bar(np.arange(len(labels)) - width, Predicted_values_by_my_model, width, label='My Model')
rects2 = ax.bar(np.arange(len(labels)), Predicted_values_by_GD, width, label='Gradient Descent')
rects3 = ax.bar(np.arange(len(labels)) + width, Predicted_values_by_SGD, width, label='Stochastic Gradient Descent')

ax.set_ylabel('Values')
ax.set_title('Predicted Values by My Model and Sklearn Model')
ax.set_xticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.legend()

# Display the bar graph for predicted values
plt.show()
```
